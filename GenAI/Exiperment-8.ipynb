```python
# pip install langchain cohere langchain-community

from langchain.prompts import PromptTemplate
from langchain_community.llms import Cohere
import getpass

# Load text file
text = open("a.txt", "r", encoding="utf-8").read()

# API Key
api_key = getpass.getpass("Enter Cohere API Key: ")

# Load model
llm = Cohere(cohere_api_key=api_key, model="command")

# Prompt template
template = """
You are an AI assistant.

Document:
{text}

1. Give a short summary.
2. List 3 key points.
3. Sentiment: Positive, Negative, or Neutral.
"""

prompt = PromptTemplate(input_variables=["text"], template=template)

# Generate response
response = llm.predict(prompt.format(text=text))

# Output
print("\nFormatted Output:\n")
print(response)
```
