```python
# pip install langchain cohere langchain-community wikipedia-api ipywidgets

import getpass, wikipediaapi, ipywidgets as widgets
from IPython.display import display
from langchain.prompts import PromptTemplate
from langchain_community.llms import Cohere

# API Key + Model
api_key = getpass.getpass("Enter Cohere API Key: ")
llm = Cohere(cohere_api_key=api_key, model="command")

# Load IPC content
wiki = wikipediaapi.Wikipedia(user_agent="IPCBot", language="en")
ipc = wiki.page("Indian Penal Code").text[:5000]

# Prompt
prompt = PromptTemplate(
    input_variables=["ipc", "question"],
    template="""
Use this IPC content to answer:
{ipc}

Question: {question}
Give section and explanation.
"""
)

# Response function
def ask_ipc(question):
    return llm.predict(prompt.format(ipc=ipc, question=question))

# Button action
def handle_click(b):
    print("\nAnswer:\n", ask_ipc(box.value))

# UI
box = widgets.Text(placeholder="Ask about IPC...")
button = widgets.Button(description="Ask")

button.on_click(handle_click)

display(box, button)
```
