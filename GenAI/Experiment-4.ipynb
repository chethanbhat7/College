```python
# Run once:
# pip install torch torchvision torchaudio transformers==4.41.2 gensim nltk

import torch
import gensim.downloader as api
import nltk, string
from nltk.tokenize import word_tokenize
from transformers import AutoTokenizer, AutoModelForCausalLM

nltk.download("punkt")

# Load models manually (avoids pipeline torch bug)
wv = api.load("glove-wiki-gigaword-100")

tokenizer = AutoTokenizer.from_pretrained("gpt2")
model = AutoModelForCausalLM.from_pretrained("gpt2")

prompt = "Who is king."

# Replace word
new_prompt = " ".join(
    [wv.most_similar("king", topn=1)[0][0]
     if w.lower().strip(string.punctuation) == "king" else w
     for w in word_tokenize(prompt)]
)

# Text generation function
def generate(text):
    inputs = tokenizer.encode(text, return_tensors="pt")
    outputs = model.generate(inputs, max_length=100)
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

r1 = generate(prompt)
r2 = generate(new_prompt)

print("Original:", prompt)
print("Enriched:", new_prompt)

print("\nOriginal Response:\n", r1)
print("\nEnriched Response:\n", r2)

print("\nLengths:", len(r1), len(r2))
print("Details:", r1.count("."), r2.count("."))
```
