```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Load and Select
df = pd.read_csv("Iris.csv")[["SepalLengthCm", "SepalWidthCm"]]

# 2. Statistics (Correlation & Covariance)
print(f"Correlation Matrix:\n{df.corr()}")
print(f"\nCovariance Matrix:\n{df.cov()}")

# 3. Visualization
plt.figure(figsize=(12, 5))

# Scatter Plot with Trend Line
plt.subplot(1, 2, 1)
sns.regplot(x=df.iloc[:,0], y=df.iloc[:,1])

# Heatmapy7
plt.subplot(1, 2, 2)
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", vmin=-1, vmax=1)

plt.tight_layout()
plt.show()
```
