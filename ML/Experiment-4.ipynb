```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import confusion_matrix

# 1. Load and Scale
df = pd.read_csv('IRIS.csv')
X = StandardScaler().fit_transform(df.iloc[:, :-1])
y = LabelEncoder().fit_transform(df.iloc[:, -1])
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 2. Loop through Weights and K-values
for wt in ['uniform', 'distance']:
    print(f"\n--- {wt.upper()} KNN ---")
    for k in [1, 3, 5]:
        knn = KNeighborsClassifier(n_neighbors=k, weights=wt).fit(X_train, y_train)
        y_pred = knn.predict(X_test)
        
        print(f"K={k} | Acc: {knn.score(X_test, y_test):.4f}")
        print(f"Confusion Matrix:\n{confusion_matrix(y_test, y_pred)}")
```
