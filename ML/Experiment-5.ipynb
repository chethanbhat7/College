```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 1. Load Data
df = pd.read_csv("tips.csv")
X, y = df["total_bill"].values, df["tip"].values

# 2. LWR Function (Page 15 Logic)
def lwr(x_q, X, y, tau=10):
    weights = np.exp(-((X - x_q)**2) / (2 * tau**2))
    X_b = np.c_[np.ones(len(X)), X]
    W = np.diag(weights)
    # Normal Equation: beta = (X.T * W * X)^-1 * X.T * W * y
    beta = np.linalg.pinv(X_b.T @ W @ X_b) @ (X_b.T @ W @ y)
    return np.array([1, x_q]) @ beta

# 3. Predict & Plot
x_range = np.linspace(X.min(), X.max(), 100)
y_pred = [lwr(x, X, y) for x in x_range]

plt.scatter(X, y, color='red', alpha=0.5)
plt.plot(x_range, y_pred, color='blue', lw=2)
plt.scatter(30, lwr(30, X, y), color='green', s=100) # Manual's $30 test
plt.title("Locally Weighted Regression (tau=10)")
plt.show()

print(f"Predicted Tip for $30: {lwr(30, X, y):.2f}")
```
