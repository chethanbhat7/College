```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import accuracy_score, confusion_matrix

# Load data
df = pd.read_csv("breast-cancer.csv", na_values='?').dropna(axis=1, how='all')
df.fillna(df.mean(numeric_only=True), inplace=True)

# Features and labels
if df.iloc[:, 1].dtype == object:   # B/M dataset
    X = df.iloc[:, 2:]
    y = df.iloc[:, 1].map({'B': 0, 'M': 1})
else:                               # 2/4 dataset
    X = df.iloc[:, 1:-1]
    y = df.iloc[:, -1].map({2: 0, 4: 1})

# Scale + Cluster
X = StandardScaler().fit_transform(X)
clusters = KMeans(n_clusters=2, random_state=42, n_init=10).fit_predict(X)

# Accuracy correction
acc = accuracy_score(y, clusters)
if accuracy_score(y, 1 - clusters) > acc:
    clusters = 1 - clusters
    acc = accuracy_score(y, clusters)

# PCA
X_pca = PCA(n_components=2).fit_transform(X)

# Plot
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=clusters, cmap='viridis', alpha=0.7)
plt.title("Breast Cancer Clustering using KMeans + PCA")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.colorbar(label="Cluster")
plt.show()

# Output
print("KMeans Clustering Accuracy:", round(acc * 100, 2), "%")
print("Confusion Matrix:\n", confusion_matrix(y, clusters))
```
