## NLTK

```python
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt_tab')

text = "Natural Language Processing is amazing"
tokens = word_tokenize(text)
print("Tokens:", tokens)
```

```python
print("Token index of 'Language':", tokens.index("Language") if "Language" in tokens else "Not found")
```

## SPACY

```python
import spacy
nlp = spacy.load("en_core_web_sm")
text = "Elon Musk founded SpaceX in 2002"
doc = nlp(text)
for ent in doc.ents:
    print(f"{ent.text} -> {ent.label_}")
```

## Text Blob

```python
from textblob import TextBlob
text = "I love Python. It is so powerful"
blob = TextBlob(text)
print("Sentiment:", blob.sentiment)
```

## Scikit-learn

```python
from sklearn.feature_extraction.text import TfidfVectorizer

corpus = ["NLP is amazing", "I love learning NLP", "Deep learning is powerful"]
vec = TfidfVectorizer()
X = vec.fit_transform(corpus)
print("Features:", vec.get_feature_names_out())
print("TF-IDF matrix shape:", X.shape)
```

## Transformers

```python
from transformers import pipeline
gen = pipeline('text-generation', model='distilgpt2')
text = gen('Artificial Intelligence', max_length=20, num_return_sequences=1)
print(text[0]['generated_text'])
```
