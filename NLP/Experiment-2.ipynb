```python
import nltk
from nltk import word_tokenize, ngrams
from collections import Counter

nltk.download('punkt')

text = "i love programming. programming is fun. i love coding and programming."
words = word_tokenize(text.lower())

# n-gram probabilities
def get_prob(n):
    counts = Counter(ngrams(words, n))
    total = sum(counts.values())
    return {g: c/total for g, c in counts.items()}

uni, bi, tri = get_prob(1), get_prob(2), get_prob(3)

print("Unigram:", uni)
print("Bigram:", bi)
print("Trigram:", tri)

# sentence probability
def prob(sentence, model, n):
    w = word_tokenize(sentence.lower())
    p = 1
    for g in ngrams(w, n):
        p *= model.get(g, 1e-10)  # smoothing
    return p

s = "i love programming"
print("Bigram Prob:", prob(s, bi, 2))
print("Trigram Prob:", prob(s, tri, 3))
```
