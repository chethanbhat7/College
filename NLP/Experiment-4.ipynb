```python
class TD:
    def __init__(self, s):
        self.s, self.i = s, 0

    def match(self, ch):
        if self.i < len(self.s) and self.s[self.i] == ch:
            self.i += 1
            return True
        return False

    def parse(self):
        if not self.match('a'): return False
        self.match('b')            # optional
        return self.match('c') and self.i == len(self.s)

print("Top-Down:", TD("abc").parse())
```

```python
class BU:
    def __init__(self, s):
        self.inp, self.st = list(s), []

    def parse(self):
        for ch in self.inp:
            self.st.append(ch)

            if ''.join(self.st[-2:]) == 'bc':
                self.st[-2:] = ['A']
            if ''.join(self.st[-2:]) == 'aA':
                self.st[-2:] = ['S']

        return self.st == ['S']

print("Bottom-Up:", BU("abc").parse())
```
