```python
class TDP:
    def __init__(self, s):
        self.s = s
    def parse(self):
        return self.s.startswith('a') and self.s.endswith('c') and 2 <= len(self.s) <= 3

class BUP:
    def __init__(self, s):
        self.s = s
    def parse(self):
        return self.s == "abc"

td = TDP("abc")
print("Top-Down parsing Result:", td.parse())
bup = BUP("abc")
print("Bottom-Up Parsing Result:", bup.parse())
```
