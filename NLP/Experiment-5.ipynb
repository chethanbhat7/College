```python
from collections import defaultdict
import math

# data
data = [
    ("fun, couple, love, love, comedy", "Comedy"),
    ("fast, furious, shoot, action", "Action"),
    ("couple, fly, fast, fun, fun, comedy", "Comedy"),
    ("furious, shoot, shoot, fun, action", "Action"),
    ("fly, fast, shoot, love, action", "Action")
]
test_doc = "fast, couple, shoot, fly"

# tokenize
tok = lambda t: t.lower().split(", ")

# train
def train(data):
    wc, cc, vocab = defaultdict(lambda: defaultdict(int)), defaultdict(int), set()
    for text, label in data:
        cc[label] += 1
        for w in tok(text):
            wc[label][w] += 1
            vocab.add(w)
    return wc, cc, len(vocab)

# classify
def predict(text, wc, cc, V):
    total = sum(cc.values())
    scores = {
        c: math.log(cc[c]/total) + sum(
            math.log((wc[c][w]+1)/(sum(wc[c].values())+V))
            for w in tok(text)
        )
        for c in cc
    }
    return max(scores, key=scores.get)

# run
wc, cc, V = train(data)
print("Predicted:", predict(test_doc, wc, cc, V))
```
