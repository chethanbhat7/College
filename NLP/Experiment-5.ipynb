```python
from collections import defaultdict
import math

data = [
    ("fun couple love love comedy", "Comedy"),
    ("fast furious shoot action", "Action"),
    ("couple fly fast fun fun comedy", "Comedy"),
    ("furious shoot shoot fun action", "Action"),
    ("fly fast shoot love action", "Action")
]

test_doc = "fast couple shoot fly"

# train
wc = defaultdict(lambda: defaultdict(int))
cc = defaultdict(int)
vocab = set()

for text, label in data:
    cc[label] += 1
    for w in text.split():
        wc[label][w] += 1
        vocab.add(w)

V = len(vocab)

# predict
def predict(text):
    total = sum(cc.values())
    scores = {}

    for c in cc:
        score = math.log(cc[c] / total)  # prior
        total_words = sum(wc[c].values())

        for w in text.split():
            score += math.log((wc[c][w] + 1) / (total_words + V))  # likelihood

        scores[c] = score

    return max(scores, key=scores.get)

print("Predicted:", predict(test_doc))
```
