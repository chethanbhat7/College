```python
import nltk
from nltk.corpus import brown, inaugural, reuters, udhr
from nltk.probability import FreqDist, ConditionalFreqDist
from nltk.tag import UnigramTagger
from nltk.corpus import PlaintextCorpusReader

# Download required datasets (run once)
for pkg in ['brown', 'inaugural', 'reuters', 'udhr', 'punkt', 'averaged_perceptron_tagger']:
    nltk.download(pkg)

# 1. Explore corpora
print("Brown Categories:", brown.categories())
print("Inaugural:", inaugural.words()[:10])
print("Reuters:", reuters.words()[:10])
print("UDHR:", udhr.words('English-Latin1')[:10])

# 2. Custom corpus
def custom_corpus(path):
    corpus = PlaintextCorpusReader(path, '.*\.txt')
    print("Files:", corpus.fileids())
    print("Words:", corpus.words()[:20])

# custom_corpus("path/to/folder")

# 3. Conditional Frequency
cfd = ConditionalFreqDist(
    (cat, w.lower())
    for cat in brown.categories()
    for w in brown.words(categories=cat)
)
print("News words:", cfd['news'].most_common(10))

# 4. Frequent nouns
nouns = [w for w, t in brown.tagged_words() if t.startswith('NN')]
print("Top nouns:", FreqDist(nouns).most_common(10))

# 5. Word properties
word_dict = {
    "AI": "Artificial Intelligence",
    "NLP": "Natural Language Processing",
    "ML": "Machine Learning"
}
print("NLP:", word_dict["NLP"])

# 6. Unigram Tagger
sents = brown.tagged_sents(categories='news')
train, test = sents[:4000], sents[4000:]
tagger = UnigramTagger(train)
print("Accuracy:", round(tagger.accuracy(test), 4))

# 7. Word Segmentation
def segment(text, vocab):
    res = []
    while text:
        for i in range(len(text), 0, -1):
            if text[:i] in vocab:
                res.append(text[:i])
                text = text[i:]
                break
        else:
            res.append(text[0])
            text = text[1:]
    return res

vocab = set(w.lower() for w in brown.words())
print("Segmented:", segment("thisisatextwithoutspaces", vocab))
```
